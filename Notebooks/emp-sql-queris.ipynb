```python
%sql
SELECT DEPTNO, 
       MAX(SAL) AS MAX_SAL, 
       MIN(SAL) AS MIN_SAL, 
       AVG(SAL) AS AVG_SAL
FROM emp
GROUP BY DEPTNO;
```

```python
%sql
SELECT EMPNO, ENAME, SAL * 12 AS ANNSAL
FROM emp;
```

```python
%sql
SELECT EMPNO, SAL, JOB FROM emp ORDER BY SAL DESC LIMIT 5;
```
