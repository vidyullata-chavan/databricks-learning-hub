```python

from pyspark.sql.types import StructType, StructField, IntegerType, StringType
schema = StructType([
    StructField("deptno", IntegerType(), True),
    StructField("dname", StringType(), True),
    StructField("loc", StringType(), True)
])
path = "/Volumes/workspace/default/vid_volume/vid_dir/mydept.csv"
df = (spark.read.format("csv")
      .option("header", "true")
      .schema(schema)
      .load(path))
	  
df.printSchema()
df.show()
```

```python
df.write.saveAsTable('default.mydept_from_csv')
```

```python
%sql
create or replace table mydept_from_csv_deepclone deep clone mydept_from_csv
```
